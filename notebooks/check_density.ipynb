```python
from learnable_typewriter.data.dataset import UniDataset
```

```python
dataset = UniDataset(path = "/home/vlachoum/PROJECT/learnable-scriber/datasets/textualis_formata/", sep = '', space = ' ', p = 0, supervised = True, height = 96, alias = 'textualis-all', split = 'train', )
```

```python
import seaborn as sns
import pandas as pd
import os

hist = []
for i in range(len(dataset)):
    img, label = dataset[i]
    path = dataset.data[i][0]
    path = os.path.split(path)[1]
    
    w = img.width
    h = img.height
    y = label
    #print((w*4)/h, len(y))
    script = '_'.join(path.split('_')[:1])
        
    hist.append(((w/h)/len(y), script))

hist = pd.DataFrame(hist, columns=['P/N', 'script'])
sns.histplot(data=hist , x="P/N", hue="script")
```
