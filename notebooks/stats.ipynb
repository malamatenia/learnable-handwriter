```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import unicodedata
from matplotlib.font_manager import FontProperties
from collections import Counter

# Load the JSON data
with open('datasets/south_north_textualis/annotation.json', 'r') as file:
    data = json.load(file)

# Initialize DataFrames for each script class
df_northern = pd.DataFrame({k: v for k, v in data.items() if v['script'] == 'Northern_Textualis'}).T
df_southern = pd.DataFrame({k: v for k, v in data.items() if v['script'] == 'Southern_Textualis'}).T

# Combine all text annotations into a single string for each class
combined_text_northern = ' '.join(df_northern['label'])
combined_text_southern = ' '.join(df_southern['label'])

# Normalize the combined text to NFD (Normalization Form D) for each class
normalized_text_northern = unicodedata.normalize('NFD', combined_text_northern)
normalized_text_southern = unicodedata.normalize('NFD', combined_text_southern)

# Remove spaces and newline characters
normalized_text_northern = normalized_text_northern.replace(' ', '').replace('\n', '')
normalized_text_southern = normalized_text_southern.replace(' ', '').replace('\n', '')

# Calculate character frequencies for each class
char_frequencies_northern = pd.Series(list(normalized_text_northern)).value_counts().sort_values(ascending=False)
char_frequencies_southern = pd.Series(list(normalized_text_southern)).value_counts().sort_values(ascending=False)

# Function to plot histogram with custom font
def plot_histogram(character_counter, title, threshold=10):
    sorted_characters = sorted(character_counter.items(), key=lambda x: x[1], reverse=True)
    above_threshold = [(char, count) for char, count in sorted_characters if count >= threshold]

    characters, counts = zip(*above_threshold)

    # Specify the path to the Junicode font file
    font_path = 'junicode-1.002/Junicode.ttf'  # Replace with the actual path to the Junicode font

    # Register the font with Matplotlib
    FontProperties(fname=font_path)

    # Specify Junicode as the font
    plt.rcParams['font.family'] = 'Junicode'

    # Plot the histogram
    plt.figure(figsize=(12, 6))
    plt.bar(characters, counts)
    plt.xlabel('Characters')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xticks(rotation=10)
    plt.tight_layout()

    # Save the figure
    plt.savefig(f'{title.lower().replace(" ", "_")}_character_frequencies.png')

# Plot histogram with custom font for each class
plot_histogram(char_frequencies_northern, 'Northern Textualis Character Frequencies', threshold=10)
plot_histogram(char_frequencies_southern, 'Southern Textualis Character Frequencies', threshold=10)

# Create DataFrames with characters and their frequencies
df_characters_northern = pd.DataFrame(char_frequencies_northern, columns=['Frequency'])
df_characters_southern = pd.DataFrame(char_frequencies_southern, columns=['Frequency'])

# Add a column for the character in each DataFrame
df_characters_northern['Character'] = df_characters_northern.index
df_characters_southern['Character'] = df_characters_southern.index

# Add columns for the number of lines and characters in each script class
df_statistics_northern = pd.DataFrame({
    'Number of Lines': [df_northern.shape[0]],
    'Number of Characters': [len(normalized_text_northern)],
}, index=['Northern_Textualis'])

df_statistics_southern = pd.DataFrame({
    'Number of Lines': [df_southern.shape[0]],
    'Number of Characters': [len(normalized_text_southern)],
}, index=['Southern_Textualis'])

# Save DataFrames to CSV files
#df_characters_northern.to_csv('df_characters_northern.csv', index=False)
#df_characters_southern.to_csv('df_characters_southern.csv', index=False)

#df_statistics_northern.to_csv('df_statistics_northern.csv')
#df_statistics_southern.to_csv('df_statistics_southern.csv')
```

```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import unicodedata
import matplotlib.font_manager as font_manager

# Load the JSON data
with open('datasets/south_north_textualis/annotation.json', 'r') as file:
    data = json.load(file)

# Initialize a DataFrame to store the text annotations
df = pd.DataFrame(data).T

# Combine all text annotations into a single string
combined_text = ' '.join(df['label'])

# Normalize the combined text to NFD (Normalization Form D)
normalized_text = unicodedata.normalize('NFD', combined_text)

# Remove spaces and newline characters
normalized_text = normalized_text.replace(' ', '').replace('\n', '')

# Calculate character frequencies
char_frequencies = pd.Series(list(normalized_text)).value_counts().sort_values(ascending=False)

# Function to plot histogram with custom font
def plot_histogram(character_counter, title, threshold=10):
    sorted_characters = sorted(character_counter.items(), key=lambda x: x[1], reverse=True)
    above_threshold = [(char, count) for char, count in sorted_characters if count >= threshold]
    
    characters, counts = zip(*above_threshold)
    
    # Specify the path to the Junicode font file
    font_path = '.media/Junicode.ttf'  # Replace with the actual path to the Junicode font

    # Register the font with Matplotlib
    font_manager.fontManager.addfont(font_path)

    # Specify Junicode as the font
    plt.rcParams['font.family'] = 'Junicode'

    # Plot the histogram
    plt.figure(figsize=(12, 6))
    plt.bar(characters, counts)
    plt.xlabel('Characters')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xticks(rotation=10)
    plt.tight_layout()

    # Save the figure
    plt.savefig('character_frequencies.png')

# Plot histogram with custom font
plot_histogram(char_frequencies, 'Character Frequencies for Northern/Southern Textualis dataset', threshold=10)
```
